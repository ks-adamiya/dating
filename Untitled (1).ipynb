Использую ANOVA, потому что при больших размерах групп (n > 5000) центральная предельная теорема гарантирует нормальность средних, а ANOVA устойчива к небольшим нарушениям предпосылок — это надёжный и интерпретируемый способ проверить общий дисбаланс по возрасту между группами.

```python
grouped = users_AB.groupby('group')['age'].apply(list)

# ANOVA
f, p = stats.f_oneway(*grouped)
print(f"p-value: {p:.4f}")
print("Дисбаланс!" if p < 0.05 else "Группы сбалансированы по возрасту")
```

Для анализа используем хи-квадрат — стандартный, корректный и интерпретируемый метод для проверки баланса категориальных признаков (таких как пол) в A/B-тестах.

```python
contingency = pd.crosstab(users_AB['group'], users_AB['gender'])
chi2, p, _, _ = stats.chi2_contingency(contingency)
print(f"Chi-square test: p = {p:.4f}")
print("Дисбаланс по полу!" if p < 0.05 else "Группы сбалансированы по полу")
```

Для проверки баланса также используем Хи-квадрат, потому что страна — категориальная переменная, а χ²-тест проверяет, одинаково ли распределены категории между группами.

```python
contingency = pd.crosstab(users_AB['group'], users_AB['country'])

# Хи-квадрат тест
chi2, p, dof, expected = chi2_contingency(contingency)
print(f"χ² = {chi2:.2f}, p-value = {p:.4f}")
print(f"{'Дисбаланс по странам!' if p < 0.05 else 'Группы сбалансированы по странам'}")
```

## Оценка результатов эксперимента

Для оценки результатов эксперимента рассчитаем CR, ARPU и ARPPU по каждой группе.

1. Конверсия в платящего пользователя. CR = (число сконвертировавших) / (всего пользователей)

```python
converted_users = transactions_AB['uid'].drop_duplicates()
users_AB['converted'] = users_AB['uid'].isin(converted_users).astype(int)

summary = users_AB.groupby('group').agg(
    total=('uid', 'count'),
    converted=('converted', 'sum')
).reset_index()
summary['cr'] = summary['converted'] / summary['total']
print("CR по группам:")
print(summary.round(4))

print("\nПопарные сравнения (χ² + Bonferroni):")
results = []

for g1, g2 in combinations(summary['group'], 2):
    n1 = summary.loc[summary['group'] == g1, 'total'].iloc[0]
    x1 = summary.loc[summary['group'] == g1, 'converted'].iloc[0]
    n2 = summary.loc[summary['group'] == g2, 'total'].iloc[0]
    x2 = summary.loc[summary['group'] == g2, 'converted'].iloc[0]
    
    table = [[x1, n1 - x1],
             [x2, n2 - x2]]
    
    chi2, p, _, _ = chi2_contingency(table, correction=False)  
    
    diff = (x1 / n1) - (x2 / n2)
    results.append({
        'pair': f"{g1} vs {g2}",
        'diff': diff,
        'p_value': p
    })

# Применяем поправку Бонферрони
res_df = pd.DataFrame(results)
res_df['p_adj'] = res_df['p_value'] * 3  
res_df['significant'] = res_df['p_adj'] < 0.05

print(res_df[['pair', 'diff', 'p_value', 'p_adj', 'significant']].round(4))
```

*Для сравнения CR в группах был использован χ²-тест, потому что он статистически корректен для сравнения пропорций (конверсии) в больших выборках.
Применяется поправка Бонферрони, потому что одновременное проведение нескольких тестов без коррекции искусственно завышает вероятность ложноположительного результата.
Вместе эти методы обеспечивают надёжную и строгую проверку гипотез о различиях в конверсии между группами.*

2. ARPU (Average revenue per user) - оценка средней выручки одного пользователя

```python
pairs = [('test', 'control_1'), ('test', 'control_2')]

print("ARPU: Welch's t-test (с поправкой Бонферрони)")
for i, (g1, g2) in enumerate(pairs, 1):
    data1 = users_AB.query(f'group == "{g1}"')['total_revenue']
    data2 = users_AB.query(f'group == "{g2}"')['total_revenue']
    
    t_stat, p = stats.ttest_ind(data1, data2, equal_var=False)
    p_adj = min(p * 2, 1.0)  # Бонферрони для 2 тестов
    
    print(f'p-value "{g1}-{g2}": {p:.4f} → adj.p = {p_adj:.4f}')
```

```python
# Удалим выбросы из транзакций
transactions_clean = transactions_AB[transactions_AB['revenue'] <= upper_bound].copy()

# Пересчитаем total_revenue на пользователя без выбросов
revenue_clean = transactions_clean.groupby('uid')['revenue'].sum().reset_index()
revenue_clean.columns = ['uid', 'total_revenue_clean']

# Присоединим к users_AB
users_clean = users_AB[['uid', 'group']].merge(revenue_clean, on='uid', how='left')
users_clean['total_revenue_clean'] = users_clean['total_revenue_clean'].fillna(0)

# Сравним ARPU между группами (Welch's t-test)
print("ARPU без выбросов (Welch's t-test):")
p_values = []
for g1, g2 in pairs:
    data1 = users_clean.query(f'group == "{g1}"')['total_revenue_clean']
    data2 = users_clean.query(f'group == "{g2}"')['total_revenue_clean']
    
    _, p = stats.ttest_ind(data1, data2, equal_var=False)
    p_values.append(p)

# Поправка Бонферрони: 
p_adj = [min(p * 2, 1.0) for p in p_values]

for (g1, g2), p, p_a in zip(pairs, p_values, p_adj):
    print(f"  {g1} vs {g2}: p = {p:.4f} → adj.p = {p_a:.4f} {'True' if p_a < 0.05 else 'False'}")
```

Для сравнения был использован Welch’s t-test, потому что он корректно сравнивает средние значения при неравных дисперсиях — типичной ситуации для метрик выручки.
Применена поправка Бонферрони, потому что одновременное проведение нескольких тестов без коррекции искусственно завышает вероятность ложных выводов.
Эти методы обеспечивают статистически строгую оценку различий в ARPU между группами.

*Вывод: если выбросы не убирать, то p-value > 0,05, т.е.  нулевую гипотезу не отвергаем, проведенные изменения не повлияли на ARPU
если выбросы убрать, то p-value < 0,05, т.е. нулевую гипотезу отвергаем, проведенные изменения повлияли на ARPU*

3. ARPPU (Average revenue per paying user) - оценка средней выручки одного платящего пользователя премиум-подписки

```python
print("ARPPU: Welch's t-test (с поправкой Бонферрони)")

# Фильтруем только платящих пользователей
payers_original = users_AB[users_AB['total_revenue'] > 0]

for g1, g2 in pairs:
    data1 = payers_original.query(f'group == "{g1}"')['total_revenue']
    data2 = payers_original.query(f'group == "{g2}"')['total_revenue']
    
    # Если в одной из групп нет платящих — пропускаем
    if len(data1) == 0 or len(data2) == 0:
        print(f'p-value "{g1}-{g2}": недостаточно данных')
        continue
        
    t_stat, p = stats.ttest_ind(data1, data2, equal_var=False)
    p_adj = min(p * 2, 1.0)  # Бонферрони для 2 тестов
    
    print(f'p-value "{g1}-{g2}": {p:.4f} → adj.p = {p_adj:.4f}')
```

```python
# 2. УДАЛЯЕМ ВЫБРОСЫ И СЧИТАЕМ ARPPU БЕЗ НИХ

# Фильтруем только платящих (total_revenue_clean > 0)
payers_clean = users_clean[users_clean['total_revenue_clean'] > 0]

# Сравним ARPPU между группами (Welch's t-test)
print("ARPPU без выбросов (Welch's t-test + Bonferroni):")
p_values = []
valid_pairs = []

for g1, g2 in pairs:
    data1 = payers_clean.query(f'group == "{g1}"')['total_revenue_clean']
    data2 = payers_clean.query(f'group == "{g2}"')['total_revenue_clean']
    
    if len(data1) == 0 or len(data2) == 0:
        print(f"  {g1} vs {g2}: недостаточно платящих пользователей")
        continue
        
    _, p = stats.ttest_ind(data1, data2, equal_var=False)
    p_values.append(p)
    valid_pairs.append((g1, g2))

# Поправка Бонферрони только для валидных пар
if p_values:
    p_adj = [min(p * len(valid_pairs), 1.0) for p in p_values]
    for (g1, g2), p, p_a in zip(valid_pairs, p_values, p_adj):
        print(f"  {g1} vs {g2}: p = {p:.4f} → adj.p = {p_a:.4f} {'True' if p_a < 0.05 else 'False'}")
```

Обоснование для применения данного статистического теста то же, что и для ARPU.

*Вывод: И с выбросами, и без нет оснований отвергнуть нулевую гипотезу. Проведенные изменения не повлияли на ARPPU.*
